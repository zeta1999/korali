```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

import numpy as np

file = '../data/results.dat'
gp = np.loadtxt(file,delimiter=' ')

file = '../data/sincos1d_train.dat'
data = np.loadtxt(file,delimiter=' ')

gp = gp[ np.argsort(gp[:,0]) ]



fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot( data[:,0], data[:,1], '.', color='blue',label='Data', linewidth=2)
ax.plot( gp[:,0], gp[:,1], color='red',label='GP mean', linewidth=2 )
ax.fill_between(gp[:,0], gp[:,1]+np.sqrt(gp[:,2]), gp[:,1]-np.sqrt(gp[:,2]), facecolor='red', alpha=0.5 ,label='GP Standard Dev.')
plt.legend()
plt.show()

```
